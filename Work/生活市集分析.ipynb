```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
```

```python
data = pd.read_csv('./Data/生活市集.csv')
data.head()
```

```python
data.sort_values(by=['当前价格', '已抢购件数'])
```

```python
small_count = data.groupby(by='小类')['已抢购件数'].mean()
s_c = small_count.sort_values(ascending=False)
s_c
```

```python
# 按类别统计小类产品的数量
s_c.plot(kind='bar', figsize=(50,8))
plt.grid()
plt.savefig('./Picture/类别统计.png')
plt.show()
```

```python
# 以类别统计小类产品的销量
data.groupby(by='小类').sum()
```

```python
# 大类统计
big_count = data.groupby(by='大类').count()
big_count
```

```python
big = big_count.drop(['skuid(拼接用)','商品名称','描述','品级','是否明星商品','评分','评分数'], axis=1)
big
```

```python
d_count = data.groupby('大类')['当前价格','已抢购件数'].mean()
d_plt = d_count.sort_values(by=['已抢购件数'], ascending=False)
d_plt
```

```python
d_plt['已抢购件数'].plot(kind='bar', figsize=(20,8))
plt.show()
```
