```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
```

# 合并

```python
mouth12 = pd.read_excel('./Data/订单/12月订单.xls')
mouth_1 = pd.read_excel('./Data/订单/1月订单.xls')
```

```python
mouth_1.head()
```

```python
# 合并两个表
result = pd.concat([mouth_1, mouth12], sort=False)
```

```python
# 处理\t指标符
result['下单时间'] = [ti.strip() for ti in result['下单时间']]
```

```python
result.head()
```

```python
# 新建一列 日期
result['日期'] = [day.split(' ')[0] for day in result['下单时间']]
```

```python
result.head()
```

```python
# 以日期和产品名称分类，以每个订单号为一单聚合统计
count = result.groupby(by=['日期','产品名称'])['订单号'].count()
```

```python
# 以日期升序排序
count_2 = count.sort_index()
```

```python
# 将series转为DataFrame
count_2df = count_2.to_frame()
```

```python
# 先存起来
count_2df.to_excel('./Data/countByDay.xlsx')
```

```python
count_2df
```

```python
# 读一下
byDay = pd.read_excel('./Data/countByDay.xlsx')
```

```python
# 判断是否存在缺省值
np.all(pd.notnull(byDay))
```

```python
# 将缺省值填充为上一行的值
byDay2 = byDay.fillna(method='ffill', axis=0)
```

```python
byDay2.to_excel('./Data/all.xlsx', index=False)
```

```python
np.all(pd.notnull(byDay2))
```

```python
# 重新生成时间序列索引
timeIndex = pd.date_range('12/5/2019','1/15/2020')
timeIndex
```

```python
# 建立一个38行（天数），102列的0（商品）
zeros = np.zeros([timeIndex.shape[0], goodsList.shape[0]])
zeros
```

```python
# 创建一个DataFrame
tempdf = pd.DataFrame(zeros, columns=goodsList, index=timeIndex)
tempdf
```

```python
# 添加日期列
tempdf['日期'] = timeIndex
tempdf.head()
```

```python
# 然后将日期设置为索引
tempdf.set_index(timeIndex)
tempdf
```

```python
for index,names,num in all.itertuples():
    tempdf.ix[index, names] = num
```

```python
tempdf.to_excel('./Data/订单/1-17.xlsx')
```

# 商家分布

```python
mall = pd.read_excel('./Data/订单/地区分布.xls')
```

```python
mall.head()
```

```python
mall.columns
```

```python
# 以卖家地区分类统计
mall_count = mall.groupby(by='卖家地区')['卖家地区'].count()
```

```python
by_fa = mall_count.sort_values(ascending=False)
```

```python
by_fa.plot.pie(figsize=(15,15),
               autopct='%.2f%%',
               textprops={'fontsize':14,'color':'black'},
               labeldistance=0.8,
               pctdistance=1.1)
plt.savefig('./Data/订单/地区分布.png')
plt.show()
```
