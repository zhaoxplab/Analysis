### 均匀分布

```python
np.random.rand
# 返回[0.0, 1.0]内的一组均匀分布的数

np.random.uniform(low=0.0, high=1.0, size=None)
# 从一个均匀分布(low, high)中随机采样，定义域“左闭右开”
# low：采样下界，float类型，默认值0；high：采样上界，float类型，默认值1；size：输出样本数目，为int或tuple

np.random.randint(low,high=None,size=None,dtype='')
# 从一个均匀分布中随机采样，生成一个整数或n维整数数组
# 若high不为None，取（low，high）之间的随机整数，否则取值（0，low）之间的随机整数
```

```python
# 生成均匀分布的随机数
x2 = np.random.uniform(-1, 1, 10000000)
x2
plt.figure(figsize=(10,10), dpi=100)
plt.hist(x=x2, bins=1000)
plt.show()
```

## 三元运算符

```python
sc = np.random.randint(40, 100, (10,5))
sc
```

```python
# 判断前四名学生 前四门课程中成绩大于60的置为1，否则为0
tem = sc[:4, :4]
np.where(tem > 60, 1, 0)
```

```python
# 大于60且小于90置为1，否则为0
np.where(np.logical_and(tem > 60, tem < 90), 1, 0)
```

```python
# 大于90或小于50置为1，否则为0
np.where(np.logical_or(tem > 90, tem < 50), 1, 0)
```

```python
# 
np.where(np.logical_not(tem > 80), 0, 1)
```

```python
tem
```

### 广播机制

```python
arra = np.array([[0],[1],[2],[3]])
arrb = np.array([1,2,3])
```

```python
arra.shape
```

```python
# 两个不懂维度的数组进行运算，按照广播机制会对arr1和arr2进行扩展，使得arra和arrb都变成4行3列
arra + arrb
```

广播机制要求：
- 数组某一维度等长
- 其中一个数组某一维度为1
> 广播机制需要扩展维度小的数组，使得它与维度最大的数组的shape值相同，以便使用元素级函数或者运算符进行运算
