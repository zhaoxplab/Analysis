# 实现基础绘图功能

```python
import matplotlib.pyplot as plt
import random
from pylab import mpl
# 设置显示中文字体
mpl.rcParams["font.sans-serif"] = ["SimHei"]
# 设置正常显示符号
mpl.rcParams["axes.unicode_minus"] = False
```

## 绘制基本图像

```python
# 准备数据
x = range(60)
y = [random.uniform(15, 18) for i in x]

# 创建画布
plt.figure(figsize=(20,8), dpi=100)
# 绘制图像
plt.plot(x, y)
plt.show()
```

## 实现其他功能

```python
# 准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
# 创建画布
plt.figure(figsize=(20,8), dpi=100)
# 绘制图像
plt.plot(x, y_shanghai)

# 添加x，y轴刻度
# 设置x，y轴刻度
x_ticks = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)
# 修改x，y轴坐标显示
# plt.xticks(x_ticks[::5]) # 坐标刻度不可以直接通过字符串进行修改
plt.xticks(x[::5], x_ticks[::5])
plt.yticks(y_ticks[::5])

# 添加x，y轴描述信息及标题
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('中午11点0分到12点0分温度变化图示', fontsize=20)

# 添加网格显示
plt.grid(True, linestyle='--', alpha=1)
# 保存图片到指定路径，要在show()之前，因为show()会释放figure资源
plt.savefig('./Pic/中午11点0分到12点0分温度变化图示.png')
# 图像显示
plt.show()
```

# 在一个坐标系中绘制多个图像

```python
# 0.准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

# 1.创建画布
plt.figure(figsize=(20,8), dpi=100)

# 2.绘制图像
plt.plot(x, y_shanghai, label='上海')
# 多次plot可以绘制多条折线
plt.plot(x, y_beijing, color='r', linestyle='--', label='北京')

# 2.1添加x，y轴刻度
# 设置x，y轴刻度
x_ticks = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)
# 2.2修改x，y轴坐标显示
# plt.xticks(x_ticks[::5]) # 坐标刻度不可以直接通过字符串进行修改
plt.xticks(x[::5], x_ticks[::5])
plt.yticks(y_ticks[::5])
# 2.3添加网格显示
plt.grid(True, linestyle='--', alpha=1)
# 2.4添加x，y轴描述信息及标题
plt.xlabel('时间')
plt.ylabel('温度')
plt.title('中午11点0分到12点0分温度变化图示', fontsize=20)
# 显示图例，loc为位置，best自动选择最好的位置
plt.legend(loc='best')

# 保存图片到指定路径，要在show()之前，因为show()会释放figure资源
plt.savefig('./Pic/中午11点0分到12点0分温度变化图示.png')
# 图像显示
plt.show()
```

# 多个坐标系实现绘图

```python
# 0.准备数据
x = range(60)
y_shanghai = [random.uniform(15, 18) for i in x]
y_beijing = [random.uniform(1, 3) for i in x]

# 1.创建画布
# plt.figure(figsize=(20,8), dpi=100)
# fig是图，axes是对象
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(20,8), dpi=60)

# 2.绘制图像，多次plot可以绘制多条折线
axes[0].plot(x, y_shanghai, label='上海')
axes[1].plot(x, y_beijing, color='r', linestyle='--', label='北京')

# 2.1添加x，y轴刻度
# 设置x，y轴刻度
x_ticks = ['11点{}分'.format(i) for i in x]
y_ticks = range(40)

# 2.2修改x，y轴坐标显示
# plt.xticks(x[::5], x_ticks[::5])
# plt.yticks(y_ticks[::5])
axes[0].set_xticks(x[::5])
axes[0].set_yticks(y_ticks[::5])
axes[0].set_xticklabels(x_ticks[::5])
axes[1].set_xticks(x[::5])
axes[1].set_yticks(y_ticks[::5])
axes[1].set_xticklabels(x_ticks[::5])

# 2.3添加网格显示
axes[0].grid(True, linestyle='--', alpha=1)
axes[1].grid(True, linestyle='--', alpha=1)

# 2.4添加x，y轴描述信息及标题
# plt.xlabel('时间')
# plt.ylabel('温度')
# plt.title('中午11点0分到12点0分温度变化图示', fontsize=20)
axes[0].set_xlabel('时间')
axes[0].set_ylabel('温度')
axes[0].set_title('中午11点0分到12点0分温度变化图示', fontsize=20)
axes[1].set_xlabel('时间')
axes[1].set_ylabel('温度')
axes[1].set_title('中午11点0分到12点0分温度变化图示', fontsize=20)
# 显示图例，loc为位置，best自动选择最好的位置
axes[0].legend(loc='best')
axes[1].legend(loc='best')

# # 保存图片到指定路径，要在show()之前，因为show()会释放figure资源
plt.savefig('./Pic/中午11点0分到12点0分温度变化图示.png')
# 图像显示
plt.show()
```

# 折线图应用举例

```python
import numpy as np

# 准备数据
x = np.linspace(-10, 10, 1000)
y = np.sin(x)

# 创建画布
plt.figure(figsize=(20,8), dpi=100)

# 绘制函数图像
plt.plot(x, y)
# 添加网格显示
plt.grid(True, linestyle='--', alpha=1)

# 显示图像
plt.show()
```
